# Question 1 — Basic GroupBy Operations

## Dataset

`imdb-top-1000.csv`

## Problem Statement

Load the dataset into a DataFrame named `movies`.

### Requirements

1. Group the dataset by `Genre`.
2. Find the total number of movies in each genre.
3. Find the average IMDb rating of each genre.
4. Find the minimum IMDb rating of each genre.
5. Find the maximum IMDb rating of each genre.
6. Sort genres by average IMDb rating in descending order.
7. Display the top 10 genres.

### Example Output

```text
Genre

Western      8.30
Crime        8.01
Drama        7.98
...
```

````python
import numpy as np
import pandas as pd
````

````python
movies = pd.read_csv("imdb-top-1000.csv")
movies
````

````python
genre = movies.groupby("Genre")
genre
````

````python
genre['Series_Title'].count()
````

````python
genre['IMDB_Rating'].mean()
````

````python
genre['IMDB_Rating'].min()
````

````python
genre['IMDB_Rating'].max()
````

````python
genre['IMDB_Rating'].mean().sort_values(ascending=False)
````

````python
genre['IMDB_Rating'].mean().sort_values(ascending=False).head(10)
````

# Question 3 — IPL GroupBy Analysis

## Dataset

`deliveries.csv`

## Problem Statement

Load the dataset into a DataFrame named `ipl`.

### Requirements

1. Find total runs scored by every batsman.
2. Find total runs conceded by every bowler.
3. Find the number of balls bowled by every bowler.
4. Find the total number of fours hit by every batsman.
5. Find the total number of sixes hit by every batsman.
6. Display the top 10 run scorers.
7. Display the top 10 wicket-taking bowlers (ignore missing dismissals).

### Example Output

```text
Top Run Scorers

Virat Kohli      7263
Rohit Sharma     6211
Shikhar Dhawan   6170
...
```

````python
ipl = pd.read_csv("deliveries.csv")
ipl
````

````python
batsman = ipl.groupby('batsman')
batsman
````

````python
batsman['batsman_runs'].count().sort_values(ascending=False)
````

````python
bowler = ipl.groupby('bowler')
bowler
````

````python
bowler['batsman_runs'].count().sort_values(ascending=False)
````

````python
bowler['ball'].count().sort_values(ascending=False)
````

````python
fours = ipl[ipl['batsman_runs'] == 4].groupby('batsman').size()
fours
````

````python
six = ipl[ipl['batsman_runs'] == 6].groupby('batsman').size()
six
````

````python
batsman['batsman_runs'].sum().sort_values(ascending=False).head(10)
````

````python
wickets_df = ipl.dropna(subset=['player_dismissed'])
top_10_bowlers = wickets_df.groupby('bowler').size().sort_values(ascending=False).head(10)
top_10_bowlers
````

# Question 5 — Real-World GroupBy Problems

## Dataset

`imdb-top-1000.csv`

## Problem Statement

Load the dataset into a DataFrame named `movies`.

### Requirements

1. Find the top 5 genres by total gross earnings.
2. Find the genre having the highest average IMDb rating.
3. Find the director whose movies received the highest total number of votes.
4. Find the number of movies directed by each director.
5. Find the average runtime of movies directed by each director.
6. Find the top 10 directors based on average IMDb rating (minimum 5 movies).
7. Display the top 10 directors based on total gross earnings.

### Example Output

```text
Top Grossing Genres

Action       8.52e9
Adventure    6.97e9
Drama        5.84e9
...
```

````python
movies = pd.read_csv("imdb-top-1000.csv")
movies
````

````python
genre = movies.groupby('Genre')
genre
````

````python
genre['Gross'].sum().sort_values(ascending=False).head(5)
````

````python
genre['IMDB_Rating'].mean().sort_values(ascending=False).head(1)
````

````python
director = movies.groupby('Director')
director
````

````python
director['No_of_Votes'].sum().sort_values(ascending=False).head(1)
````

````python
movies_per_director = movies.groupby('Director')['Series_Title'].count().sort_values(ascending=False)
movies_per_director
````

````python
director['Runtime'].mean().sort_values(ascending=False)
````

````python
director_stats = movies.groupby('Director').agg(
    movie_count=('Series_Title', 'count'),
    avg_rating=('IMDB_Rating', 'mean')
)
top_10_directors_by_rating = director_stats[director_stats['movie_count'] >= 5].sort_values(by='avg_rating', ascending=False).head(10)
top_10_directors_by_rating
````

````python
director['Gross'].sum().sort_values(ascending=False).head(10)
````
