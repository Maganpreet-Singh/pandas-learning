# Question 3 — Month-wise Revenue Analysis

## Dataset

`courses.csv`
`reg-month1.csv`
`reg-month2.csv`

## Problem Statement

Load all datasets.

### Requirements

1. Create a MultiIndex registration DataFrame using `keys`.
2. Convert the MultiIndex into columns using `reset_index()`.
3. Merge it with the courses DataFrame.
4. Find revenue generated in each month.
5. Find the average course price purchased in each month.
6. Find the total number of enrollments in each month.

### Example Output

```text
Revenue

November    248000
December    66100
```

````python
nov = pd.read_csv('reg-month1.csv')
dec = pd.read_csv('reg-month2.csv')
courses = pd.read_csv('courses.csv')
courses
````

````python
regs = pd.concat([nov,dec],keys=['Nov','Dec'])
regs
````

````python
regs_df = regs.reset_index().rename(columns={'level_0': 'Month'})
regs_df
````

````python
merge_df = regs_df.merge(courses,on='course_id',how='inner')
merge_df
````

````python
merge_df.groupby('Month')['price'].sum()
````

````python
merge_df.groupby('Month')['price'].mean()
````

````python
merge_df.groupby('Month')['student_id'].count()
````

## Problem Statement

Create two DataFrames.

### DataFrame 1

```text
emp_id   Name
1        Rahul
2        Aman
3        Neha
```

### DataFrame 2

```text
employee_id   Salary
1             45000
2             52000
4             38000
```

### Requirements

1. Merge both DataFrames.
2. Use `left_on` and `right_on`.
3. Perform Inner Join.
4. Perform Left Join.
5. Perform Outer Join.
6. Display missing values.

### Example Output

```text
emp_id   Name    Salary

1        Rahul   45000
2        Aman    52000
3        Neha      NaN
```

````python
name = {
    'emp_id' : [1,2,3],
    'Name' : ['Rahul','Aman','Neha']
}
name = pd.DataFrame(name)
name
````

````python
salary = {
    'employee_id' : [1,2,4],
    'Salary' : [45000,52000,38000]
}
salary = pd.DataFrame(salary)
salary
````

````python
merge_df = name.merge(salary,left_on='emp_id',right_on='employee_id')
merge_df
````

````python
inner_join = name.merge(salary,left_on='emp_id',right_on='employee_id',how='inner')
inner_join
````

````python
left_join = name.merge(salary,left_on='emp_id',right_on='employee_id',how='left')
left_join
````

````python
outer_join = name.merge(salary,left_on='emp_id',right_on='employee_id',how='outer')
outer_join
````

````python
outer_join.isna().sum()
````

# Question 8 — Real-World Merge Problems

## Dataset

`students.csv`
`courses.csv`
`reg-month1.csv`
`reg-month2.csv`

## Problem Statement

Load all datasets.

### Requirements

1. Find the student who spent the highest amount.
2. Find total revenue generated by each course.
3. Find the most popular course (highest enrollments).
4. Find students who enrolled in both months.
5. Find students who enrolled only in November.
6. Find students who enrolled only in December.
7. Find courses that were never purchased.
8. Find the average spending per student.

### Example Output

```text
Top Spending Student

Rahul     14997
```

````python
students = pd.read_csv("students.csv")
students
````

````python
courses = pd.read_csv('courses.csv')
courses
````

````python
nov = pd.read_csv('reg-month1.csv')
nov
````

````python
dec = pd.read_csv('reg-month2.csv')
dec
````

````python
regs = pd.concat([nov, dec], ignore_index=True)
regs
````

````python
full_df = regs.merge(students, on='student_id', how='inner').merge(courses, on='course_id', how='inner')
full_df
````

````python
full_df.groupby('name')['price'].sum().sort_values(ascending=False).head(1)
````

````python
full_df.groupby('course_name')['price'].sum()
````

````python
full_df.groupby('course_name')['student_id'].sum().sort_values(ascending=False)
````

````python
both_months_ids = set(nov['student_id']).intersection(set(dec['student_id']))
students_both_months = students[students['student_id'].isin(both_months_ids)]['name']
students_both_months
````

````python
nov_only_ids = set(nov['student_id']) - set(dec['student_id'])
students_nov_only = students[students['student_id'].isin(nov_only_ids)]['name']
students_nov_only
````

````python
dec_only_ids = set(dec['student_id']) - set(nov['student_id'])
students_dec_only = students[students['student_id'].isin(dec_only_ids)]['name']
students_dec_only
````

````python
purchased_course_ids = regs['course_id'].unique()
never_purchased_courses = courses[~courses['course_id'].isin(purchased_course_ids)]['course_name']
never_purchased_courses
````

````python
total_revenue = full_df['price'].sum()
unique_registered_students = full_df['student_id'].nunique()
avg_spending = total_revenue / unique_registered_students
avg_spending
````
