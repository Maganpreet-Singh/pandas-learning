# Question 1 — Creating and Exploring a DataFrame

## Problem Statement

Create the following DataFrame.

| Roll No | Name | Marks | Package |
|---------|------|-------|---------|
|101|Aman|82|6|
|102|Priya|91|10|
|103|Rahul|76|5|
|104|Neha|88|8|
|105|Karan|69|4|

Store it in a variable named `students`.

### Requirements

1. Display the complete DataFrame.
2. Display the shape of the DataFrame.
3. Display all column names.
4. Display the index.
5. Display the underlying NumPy array.
6. Display the data type of every column.
7. Display the first three rows.
8. Display the last two rows.
9. Display any three random rows.

### Example Output

```text
Shape : (5, 4)

Name

````python
import numpy as np
import pandas as pd
````

````python
students = {
    "Roll No" : [101,102,103,104,105],
    "Name" : ["Aman","Priya","Rahul","Neha","Karan"],
    "Marks" : [82,91,76,88,69],
    "Package" : [6,10,5,8,4]
}

students = pd.DataFrame(students)
students
````

````python
students.shape
````

````python
students.values
````

````python
students.dtypes
````

````python
students.head(3)
````

````python
students.tail(2)
````

````python
students.sample(3)
````

# Question 3 — Mathematical Analysis of Student Data

## Problem Statement

Using the `students` DataFrame created in Question 1, perform the following operations.

### Requirements

1. Find the sum of every numerical column.
2. Find the minimum value of every numerical column.
3. Find the maximum value of every numerical column.
4. Find the average of every numerical column.
5. Find the median.
6. Find the standard deviation.
7. Find the variance.
8. Calculate the sum of every row using the appropriate axis.

### Example Output

```text
Column-wise Sum

Marks        406

...
```

````python
column_sums = students.sum(numeric_only=True)

print("Column-wise Sum:")
print(column_sums)
````

````python
column_min = students.min(numeric_only=True)
column_min
````

1. Display the first row.

### Part B — Rename Columns

Create the following DataFrame.

| ID | Marks | Package |
|----|-------|---------|
|101|80|6|
|102|90|9|
|103|75|5|

Rename the columns as follows.

| Old Name | New Name |
|----------|----------|
|ID|Student_ID|
|Marks|Percentage|
|Package|LPA|

Make the changes permanent.

Display the updated DataFrame and print the updated column names.

### Example Output

```text
Updated Columns

Student_ID
Percentage
LPA

Updated DataFrame

Student_ID   Percentage   LPA
101          80           6
102          90           9
103          75           5
```


````python
movies = pd.read_csv("movies.csv")
movies
````

````python
movies.iloc[0]
````

````python
movies.iloc[-1]
````

````python
movies.iloc[20:31]
````

````python
movies.iloc[0:21:2]
````

````python
movies.iloc[5:36:10]
````

````python
movies.iloc[-15:]
````

````python
student = {
    "ID" : [101,102,103],
    "Marks" : [80,90,75],
    "Package" : [6,9,5]
}

students = pd.DataFrame(student)
students
````

````python
students.rename(columns={
    "ID": "Student_ID",
    "Marks": "Percentage",
    "Package": "LPA"
}, inplace=True)
students
````
