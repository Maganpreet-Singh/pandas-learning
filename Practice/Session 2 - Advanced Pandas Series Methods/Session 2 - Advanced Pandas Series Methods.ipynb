# Question 1 — Data Type Conversion and Range Analysis

## Dataset

`kohli_ipl.csv`

## Problem Statement

Load the dataset as a Pandas Series.

Requirements:

1. Display the original data type.
2. Convert the Series into `int16`.
3. Display the new data type.
4. Find all innings where Virat Kohli scored between **50 and 99** (both inclusive).
5. Display those innings.
6. Display the total number of such innings.

### Example Output

```text
Original Data Type : int64

Converted Data Type : int16

Scores Between 50 and 99

Match 15 : 58

Match 20 : 77

...

Total Innings : 64
```

````python
import numpy as np
import pandas as pd
````

````python
vk = pd.read_csv("kohli_ipl.csv",index_col="match_no").squeeze('columns')
vk
````

````python
print(f"Original Data Type : {vk.dtype}")
````

````python
vk = vk.astype('int16')
print(f"Current Data Type : {vk.dtype}")
````

````python
print("Score Between 50 and 99")
vk[vk.between(50,99)]
````

````python
print(f"Total Innings : {vk.between(50,99).sum()}")
````

# Question 2 — Cleaning Duplicate Data

## Problem Statement

Create the following Series.

```python
[10, 20, 30, 20, 50, 10, 60, 70, 30, 80]
```

Requirements:

1. Display the original Series.
2. Find which values are duplicates.
3. Count the total number of duplicate values.
4. Remove duplicate values while keeping the first occurrence.
5. Remove duplicate values while keeping the last occurrence.
6. Display the final Series in both cases.

### Example Output

```text
Duplicate Values

False

False

False

True

...

Total Duplicate Values : 3

Series After Removing Duplicates

10

20

30

50

60

70

80
```


````python
series = pd.Series([10, 20, 30, 20, 50, 10, 60, 70, 30, 80])
print("Original Series:")
print(series)
````

````python
print("Duplicate Values:")
series.duplicated()
````

````python
print(f"Total Duplicate Values : {series.duplicated().sum()}")
````

````python
print("Remove duplicate values while keeping the first occurrence:")
series.drop_duplicates(keep='first')
````

````python
print("Remove duplicate values while keeping the last occurrence:")
series.drop_duplicates(keep='last')
````

# Question 3 — Missing Value Investigation

## Problem Statement

Create the following Series.

```python
[12, None, 25, 40, None, 55, 68, None, 90]
```

Requirements:

1. Display the total size of the Series.
2. Display the count of non-missing values.
3. Find the total number of missing values.
4. Display all missing positions.
5. Create a new Series after removing missing values.
6. Replace missing values with 0.
7. Replace missing values using the mean of the Series.

### Example Output

```text
Series Size : 9

Non Missing Values : 6

Missing Values : 3

Series After Filling

12

0

25

40

0

55

68

0

90
```

````python
series = pd.Series([12, None, 25, 40, None, 55, 68, None, 90])
series
````

````python
print(f"Series Size : {series.size}")
print(f"Non Missing Values : {series.count().sum()}")
print(f"Missing Values : {series.isnull().sum()}")
````

````python
print("Series after removing missing values")
series.dropna()
````

````python
print("Replace missing values with 0")
series.fillna(0)
````

````python
print("Replace missing values with mean of the Series")
series.fillna(series.mean())
````
