# Question 1 — YouTube Subscribers Series Analysis

## Dataset

`subs.csv`

## Problem Statement

Load the dataset as a **Pandas Series**.

Requirements:

1. Display the first 10 records.
2. Display the last 10 records.
3. Display 5 random records.
4. Display the total number of records.
5. Display the Series name.
6. Display the data type.
7. Check whether all subscriber values are unique.
8. Print only the index.
9. Print only the values.

### Example Output

```text
First 10 Records

...

Last 10 Records

...

Random Records

...

Series Name : Subscribers Gained

```

````python
import numpy as np
import pandas as pd
````

````python
print("Series Name:")
print(subs.name)
````

````python
print("Series Data Type:")
print(subs.dtype)
````

````python
print("All Subscriber values are Unique:")
print(subs.is_unique)
````

````python
print("Index:")
subs.index
````

````python
print("Values:")
subs.values
````

# Question 4 — Series Attributes Explorer

## Dataset

Use all three datasets.

## Problem Statement

Load all three datasets as separate Series.

For each Series, display:

- Size
- Data Type
- Name
- Index
- Values
- Whether all values are unique

Finally answer the following:

- Which dataset has the most records?
- Which dataset contains duplicate values?
- Which dataset contains integer values?
- Which dataset uses custom indexing?

### Example Output

```text
Subscribers Dataset

Size : 365

Data Type : int64

Unique : False

-------------------------

Kohli Dataset

Size : 215

Data Type : int64

Unique : False

-------------------------

Bollywood Dataset

Size : 1500

Data Type : object

Unique : False
```


````python
subs = pd.read_csv("subs.csv").squeeze('columns')
vk = pd.read_csv("kohli_ipl.csv",index_col="match_no").squeeze('columns')
movies = pd.read_csv("bollywood.csv",index_col="movie").squeeze('columns')
````

````python
print("Subscribers Dataset")
print(f"Size : {subs.size}")
print(f"Data Type : {subs.dtype}")
print(f"Name : {subs.name}")
print(f"Index : {subs.index}")
print(f"Values : {subs.values}")
print(f"Unique : {subs.is_unique}")
print()
print("-"*20)
print()
print("Kohli Dataset")
print(f"Size : {vk.size}")
print(f"Data Type : {vk.dtype}")
print(f"Name : {vk.name}")
print(f"Index : {vk.index}")
print(f"Values : {vk.values}")
print(f"Unique : {vk.is_unique}")
print()
print("-"*20)
print()
print("Bollywood Dataset")
print(f"Size : {movies.size}")
print(f"Data Type : {movies.dtype}")
print(f"Name : {movies.name}")
print(f"Index : {movies.index}")
print(f"Values : {movies.values}")
print(f"Unique : {movies.is_unique}")
````

````python
print("Dataset has the Most Records:")
if (subs.size > vk.size) and (subs.size > movies.size):
    print("Subscribers Dataset")
elif vk.size > movies.size :
    print("Kohli Dataset")
else:
    print("Bollywood Dataset")
````

````python
print("Dataset has Duplicate Values:")
if not movies.is_unique:
    print("Bollwood Dataset")
if not vk.is_unique:
    print("Kohli Dataset")
if not subs.is_unique:
    print("Subscribers Dataset")
````

````python
print("Dataset has Integer Values:")
if movies.dtype == int:
    print("Bollwood Dataset")
if vk.dtype == int:
    print("Kohli Dataset")
if subs.dtype == int:
    print("Subscribers Dataset")
````

````python
print("Dataset has Custom Index:")
if not isinstance(movies.index, pd.RangeIndex):
    print("Bollwood Dataset")
if not isinstance(vk.index, pd.RangeIndex):
    print("Kohli Dataset")
if not isinstance(subs.index, pd.RangeIndex):
    print("Subscribers Dataset")
````
