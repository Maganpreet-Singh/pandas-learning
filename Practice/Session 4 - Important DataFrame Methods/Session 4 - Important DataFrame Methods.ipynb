# Question 4 — Ranking Players

## Dataset

`batsman_runs_ipl.csv`

## Problem Statement

Load the dataset into a DataFrame named `batsman`.

### Requirements

1. Create a new column named `Batting Rank`.
2. The batsman with the highest runs should receive Rank 1.
3. Display the top 20 ranked batsmen.
4. Display the bottom 20 ranked batsmen.
5. Find Virat Kohli's batting rank.
6. Find all players having Rank less than or equal to 10.

### Example Output

```text
Top Ranked Players

Rank    Batsman
1       Virat Kohli
2       Shikhar Dhawan
3       Rohit Sharma
...
```

````python
batsman = pd.read_csv("batsman_runs_ipl.csv")
batsman
````

````python
batsman['Batting Rank'] = batsman['batsman_run'].rank(ascending=False, method='min').astype(int)
batsman
````

````python
batsman.sort_values(by='Batting Rank')
````

````python
batsman.sort_values(by='Batting Rank').head(20)
````

````python
batsman.sort_values(by='Batting Rank').tail(20)
````

````python
batsman[batsman['batter'] == 'V Kohli']
````

````python
batsman[batsman['Batting Rank'] <= 10]
````

# Question 5 — Index Operations and Missing Value Analysis

## Dataset

`diabetes.csv`

## Problem Statement

Load the dataset into a DataFrame named `diabetes`.

### Part A — Index Operations

1. Set the first column as the DataFrame index.
2. Display the DataFrame.
3. Reset the index back to default.
4. Rename any three columns with meaningful names.
5. Display the updated column names.

---

### Part B — Missing Value Analysis

1. Count missing values in every column.
2. Count non-missing values in every column.
3. Display only the columns containing missing values.
4. Find the column having the highest number of missing values.
5. Display whether the DataFrame contains any missing values.

### Example Output

```text
Missing Values

Glucose        5
BloodPressure  3
BMI            7

Column with Maximum Missing Values

BMI
```

````python
diabetes = pd.read_csv('diabetes.csv')
diabetes
````

````python
diabetes = diabetes.set_index('Pregnancies')
diabetes
````

````python
diabetes = diabetes.reset_index()
diabetes
````

````python
diabetes = diabetes.rename(columns={
    'Glucose': 'Blood_Sugar_Level', 
    'BloodPressure': 'Blood_Pressure_mmHg', 
    'SkinThickness': 'Skin_Fold_Thickness'
})
diabetes
````

````python
diabetes.isnull().sum()
````

````python
diabetes.notnull().sum()
````

````python
diabetes[diabetes.isna()].sum()
````
